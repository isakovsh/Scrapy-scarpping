```python
df.dropna(inplace=True)
```

```python
url_pattern = re.compile(r'https?://\S+')
has_url = df['answer_text'].str.contains(url_pattern)

url_rows = df[has_url]
url_rows
```

```python
pattern='(https?://\S+)'
url_rows['link1']=df['answer_text'].str.extract(pattern)
```

```python
url_rows['link1'].sample(15)
```

```python
url_rows.to_csv("url_data.csv",index=False)
```
